```python
import pandas as pd
```

```python
#Q1 A
df = pd.read_csv("IRIS.csv")
df.head(10)
```

```python
print(df.shape)
print(df.dtypes)
print(df.describe())
```

```python
#Q1 B
df[(df["petal_length"] >  4.5) & (df["species"] == "Iris-virginica")]
```

```python
#Q1 C
df.groupby("species").agg(
    Average_sepal_length = ("sepal_length","mean"),
    Maximum_petal_width = ("petal_width","max"),
    Standard_deviation_of_sepal_width = ("sepal_width","std")
)
```

```python
#Q1 D
df = df.assign(petal_ratio = df["petal_length"] / df["petal_width"])
df.head()
```

```python
df.groupby("species")["petal_ratio"].agg("mean")
```
